```python
#filename = 'H:/THI- MAPE/Master Thesis/resources/corrected labels/005000.txt'
import os
import cv2

path1 = 'C:/Users/naugusti/Desktop/resources-for-thesis/training/labels-val/'
path2 = 'C:/Users/naugusti/Desktop/resources-for-thesis/training/yololabels/' 
path3 = 'C:/Users/naugusti/Desktop/resources-for-thesis/training/validate/'

list_line = []
list_file = []

for file_ in enumerate(os.listdir(path1)):
    filename1 = path1 + file_[1]
    filename2 = path2 + file_[1]

    image_name = file_[1].split(".")[0] + '.png'

    img = cv2.imread(path3 + image_name)
    size = img.shape

    

    with open(filename1) as file:
        list_file = []
        
        for line in file:
            list_line = []
            
            #print(line.rstrip()) 
            values = line.split() #splits the values of every line in the text file to individual values
            dw = 1. / size[1]
            dh = 1. / size[0]
            
            if values[0]!='DontCare':
                

                if values[0] =='Car':
                    label = 0
                elif values[0] =='Van':
                    label = 1
                elif values[0] =='Truck':
                    label = 2
                elif values[0] =='Pedestrian':
                    label = 3
                elif values[0] =='Person':
                    label = 4
                elif values[0] =='Cyclist':
                    label = 5
                elif values[0] =='Tram':
                    label = 6
                elif values[0] =='Misc':
                    label = 7

    
                x = (float(values[4]) + float(values[6])) / 2.0
                y = (float(values[5]) + float(values[7])) / 2.0
                w = float(values[6]) - float(values[4])
                h = float(values[7]) - float(values[5])
                x = round(x * dw,3)
                w = round(w * dw,3)
                y = round(y * dh,3)
                h = round(h * dh,3)
                list_line.extend([label,x,y,w,h])
                #print('list1\n',list1)

                #print('\nlist 1 -->',list1)
                list_file.append(list_line)
                #print('list2-->\n', list2)
                #print('\n')
            
            with open(filename2, 'w') as fp:
                for element in list_file:
                    fp.write(f'{element[0]}' + ' ' + f'{element[1]}' + ' ' + f'{element[2]}'+ ' ' + f'{element[3]}' + ' ' + f'{element[4]}')
                    fp.write('\n')

    print('Completed:',file_[1])
    print('\n')
```
